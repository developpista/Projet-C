```python
jrfrie=np.array(data3['JourFerie'])
jrfrie
```

```python
vac=np.array(data3['VacancesZoneC'])
vac
```

```python
pred=[]
for i in range(len(jrfrie)):
    if (jrfrie[i]==True and vac[i]==True) or (jrfrie[i]==False and vac[i]==True) :
        pred.append('Decarbonne')
    elif (jrfrie[i]==True and vac[i]==False):
        pred.append('Normal')
    else :
        pred.append('Carbonne')
pred            
```

```python
data3['MixProdElec']=pred
data3
```
